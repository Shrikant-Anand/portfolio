**Thresholding and Masking**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Cargamos las imágenes
img1 = cv2.imread('palmeras.jpg')
img2 = cv2.imread('flor.jpg')

# Convertimos la imagen a escala de grises
gris = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

# Umbralizamos la imagen para resaltar el minarete y las palmeras
r, thresh = cv2.threshold(gris, 60, 255, cv2.THRESH_BINARY_INV)

# Creamos la máscara con el umbral
mask = cv2.merge([thresh, thresh, thresh])

# Redimensionamos las imágenes al mismo tamaño
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Aplicamos la máscara con bitwise_and() sobre la imagen “flor.jpg”
res = cv2.bitwise_and(img2, mask)

images = np.concatenate((img1, mask, res), axis=1)

# Mostramos las imágenes
plt.imshow(images)
cv2.destroyAllWindows()
```

**Histogram**

```python

img_gris = cv2.imread('banderas.jpg',0)  #imagen en escala de grises 
img_color = cv2.imread('banderas.jpg')   #imagen en color 
cv2.imshow('Image', img_color)
cv2.waitKey(0)


#1 Calculamos el histograma con matplotlib     
plt.hist(img_gris.ravel(),256,[0,256])   


#2  Imagen en escala de grises
# Calculamos el histograma con OpenCV
hist = cv2.calcHist([img_gris], [0], None, [256], [0, 256])

# Dibujamos histograma con matplotlib
plt.plot(hist,color='b')
plt.xlim([0, 256])
plt.show()


#3  La imagen en color
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img_color],[i],None,[256],[0,256]) #calcula el histograma con la función de OpenCV
    plt.plot(histr,color = col)                        #dibuja con la función habitual de matplotlib
    plt.xlim([0,256])
plt.show()

cv2.destroyAllWindows()
```

**Histogram-2**


```python
# Creamos la imagen en negro con el mismo tamaño que la imagen en escala de grises
mask = np.zeros_like(img_gris)

#img_gris.shape  -> (519, 692)

# Ajustamos el blanco a la altura de la imagen de 200 a 519
mask[200:519, :] = 255

# Aplicamos la máscara a la imagen en escala de grises
res = cv2.bitwise_and(img_gris, mask)

cv2.imshow('Gris',img_gris)
cv2.imshow('Mask',mask)
cv2.imshow('Final',res)
cv2.waitKey(0) 

cv2.destroyAllWindows()

# Calculamos el histograma de la imagen completa
hist_full = cv2.calcHist([img_gris], [0], None, [256], [0, 256])

# Calculamos el histograma de la región enmascarada
hist_masked = cv2.calcHist([img_gris], [0], mask, [256], [0, 256])

# Dibujamos los histogramas
plt.plot(hist_full, color='blue')
plt.plot(hist_masked, color='orange')
plt.xlim([0, 256])
plt.show()

cv2.destroyAllWindows()
```

**Face Detection**

```python
#Cargamos el clasificador pre-entrenado para caras
gatos_clas=cv2.CascadeClassifier('haarcascade_frontalcatface.xml')

#Leemos la imagen y la transformamos en escala de grises
img = cv2.imread('cats.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#Llamamos al detector de gatos
gatos = gatos_clas.detectMultiScale(img_gray)

font = cv2.FONT_HERSHEY_SIMPLEX
for (i,(x,y,w,h)) in enumerate(gatos):
  img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,100,100),2)
  cv2.putText(img, f"Gato:{i}",(x, y - 10), font, 0.5,(0,255,0),2,cv2.LINE_AA)

#Mostramos imagen
cv2.imshow('Gatos',img)
cv2.imwrite('img.jpg',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
